```python
# --- Dummy Module Creation ---
# This is CRUCIAL. The pre-trained Lag-Llama checkpoint expects certain
# GluonTS module paths (like gluonts.torch.modules.loss) which might
# not exist in newer GluonTS versions or the way the checkpoint was saved.
# This code creates dummy modules and classes to satisfy torch.load.
```

```python
import sys
from types import ModuleType

# Create dummy module hierarchy
def create_dummy_module(module_path):
    """
    Create a dummy module hierarchy for the given path.
    Returns the leaf module.
    """
    parts = module_path.split('.')
    current = ''
    parent = None

    for part in parts:
        current = current + '.' + part if current else part
        if current not in sys.modules:
            module = ModuleType(current)
            sys.modules[current] = module
            if parent:
                setattr(sys.modules[parent], part, module)
        parent = current

    return sys.modules[module_path]

# Create the dummy gluonts module hierarchy
gluonts_module = create_dummy_module('gluonts.torch.modules.loss')

# Create dummy classes for the specific loss functions
class DistributionLoss:
    def __init__(self, *args, **kwargs):
        pass

    def __call__(self, *args, **kwargs):
        return 0.0

    def __getattr__(self, name):
        return lambda *args, **kwargs: None

class NegativeLogLikelihood:
    def __init__(self, *args, **kwargs):
        pass

    def __call__(self, *args, **kwargs):
        return 0.0

    def __getattr__(self, name):
        return lambda *args, **kwargs: None

# Add the specific classes to the module
gluonts_module.DistributionLoss = DistributionLoss
gluonts_module.NegativeLogLikelihood = NegativeLogLikelihood
```

```python
# --- Create GluonTS Datasets ---
def create_gluonts_dataset(df, all_series_names, target_series_names, freq):
    """
    Creates a GluonTS ListDataset from a pandas DataFrame.
    It prepares the data for LagLlama by identifying target series and dynamic real features.

    Args:
        df: Pandas DataFrame containing all relevant time series data.
        all_series_names: A list of all column names in 'df' that represent time series data
                          to be fed into the model (including targets and dynamic features).
        target_series_names: A list of column names that are the primary prediction targets.
        freq: The frequency of the time series (e.g., 'H' for hourly).

    Returns:
        ListDataset: A GluonTS dataset object, where each entry corresponds to a target series.
    """
    dataset = []
    # Iterate through each designated target series.
    for target_col in target_series_names:
        if target_col in df.columns:
            # Extract the target values for the current series.
            target_values = df[target_col].values.astype(np.float32)

            # --- CRITICAL FIX: Do NOT pass arbitrary dynamic features if the pre-trained model has a fixed input_size ---
            # The pre-trained Lag-Llama model (lag-llama.ckpt) has a fixed 'input_size' (e.g., 92).
            # It's typically trained as a univariate model (target only) and internally generates features like time features.
            # Passing a large number of 'FEAT_DYNAMIC_REAL' features (like 1209) will cause a size mismatch.
            # Therefore, we will NOT add FieldName.FEAT_DYNAMIC_REAL here for zero-shot forecasting with this checkpoint.
            feat_dynamic_real_values = None

            # Construct the GluonTS series dictionary.
            series_data = {
                FieldName.TARGET: target_values, # The series to be predicted
                FieldName.START: df.index[0], # The start timestamp of the series
                FieldName.ITEM_ID: target_col, # Unique identifier for this series (e.g., 'price' or 'carbonIntensity')
            }
            # Only add FEAT_DYNAMIC_REAL if it's explicitly generated and intended for the model,
            # and if the model's architecture (input_size) supports it.
            # For this pre-trained Lag-Llama, we assume it's not designed for arbitrary external dynamic features.
            # if feat_dynamic_real_values is not None:
            #     series_data[FieldName.FEAT_DYNAMIC_REAL] = feat_dynamic_real_values

            # A simple static categorical feature for each series (optional, but good for some models)
            series_data["feat_static_cat"] = [all_series_names.index(target_col) % 100]

            dataset.append(series_data)
        else:
            print(f"❗ Warning: Target column '{target_col}' not found in the DataFrame. Skipping its inclusion in the dataset.")
    return ListDataset(dataset, freq=freq)

# Create GluonTS datasets for training, validation, and testing.
train_dataset = create_gluonts_dataset(train_data, all_series_names, target_columns, FREQ)
val_dataset = create_gluonts_dataset(val_data, all_series_names, target_columns, FREQ)
test_dataset = create_gluonts_dataset(test_data, all_series_names, target_columns, FREQ)

print("✅ GluonTS datasets created for primary targets!")
```

```python
# --- LagLlama Prediction Function ---
def get_lag_llama_predictions(dataset, prediction_length, device_str, context_length, num_samples=100, use_rope_scaling=False):
    """
    Generates predictions using the pre-trained LagLlama model.

    Args:
        dataset: The GluonTS Dataset (e.g., test_dataset) for which to generate forecasts.
        prediction_length: The number of future time steps the model should predict.
        device_str: The device to run the model on ("cuda" for GPU or "cpu").
        context_length: The length of the historical context window the model uses.
        num_samples: The number of probabilistic samples to draw from the model's forecast.
        use_rope_scaling: Boolean, enables RoPE scaling for handling longer contexts if True.

    Returns:
        tuple: (forecasts, tss) - a tuple containing:
               - forecasts (list of Forecast objects): The generated probabilistic forecasts.
               - tss (list of Pandas Series): The actual time series data corresponding to the forecasts.
    """
    # Convert the device string to a torch.device object for PyTorch operations.
    torch_device = torch.device(device_str)

    # Load the pre-trained Lag-Llama checkpoint.
    # map_location ensures the checkpoint loads correctly regardless of the device it was saved on.
    # weights_only=False allows loading additional metadata like hyper-parameters.
    ckpt = torch.load("lag-llama.ckpt", map_location=torch_device, weights_only=False)
    # Extract model arguments from the checkpoint's hyper_parameters.
    estimator_args = ckpt["hyper_parameters"]["model_kwargs"]

    rope_scaling_arguments = None
    if use_rope_scaling:
        # RoPE scaling is used to extend the effective context length for the model.
        rope_scaling_arguments = {
            "type": "linear",
            # The factor scales RoPE embeddings. It's proportional to the ratio of
            # desired total length (context + prediction) to the model's original context.
            "factor": max(1.0, (context_length + prediction_length) / estimator_args["context_length"]),
        }
        print(f"Using RoPE scaling with factor: {rope_scaling_arguments['factor']:.2f}")

    # --- CRITICAL FIX for size mismatch: Use the input_size from the pre-trained checkpoint ---
    # The `RuntimeError` indicated the checkpoint expects `input_size=92`.
    # `estimator_args["input_size"]` from the loaded checkpoint should provide this value (92).
    # We must initialize the Estimator with this exact input_size.
    fixed_estimator_input_size = estimator_args["input_size"]
    print(f"Using fixed input_size from checkpoint: {fixed_estimator_input_size}")

    # Initialize the LagLlamaEstimator with the loaded checkpoint and configuration.
    estimator = LagLlamaEstimator(
        ckpt_path="lag-llama.ckpt",
        prediction_length=prediction_length,
        context_length=context_length,
        input_size=fixed_estimator_input_size, # Use the fixed input size from checkpoint (e.g., 92)
        n_layer=estimator_args["n_layer"],
        n_embd_per_head=estimator_args["n_embd_per_head"],
        n_head=estimator_args["n_head"],
        scaling=estimator_args["scaling"],
        time_feat=estimator_args["time_feat"],
        rope_scaling=rope_scaling_arguments if use_rope_scaling else None, # Pass None if not using scaling
        batch_size=64, # Batch size for inference, can be adjusted based on GPU memory.
        num_parallel_samples=num_samples, # Number of samples for probabilistic forecasts.
        device=torch_device, # Assign the determined PyTorch device.
    )

    # Create the Lightning module, transformation, and predictor from the estimator.
    lightning_module = estimator.create_lightning_module()
    transformation = estimator.create_transformation()
    predictor = estimator.create_predictor(transformation, lightning_module)

    # Generate evaluation predictions using the configured predictor.
    print(f"Generating {num_samples} samples for prediction length {prediction_length} hours on {device_str}...")
    forecast_it, ts_it = make_evaluation_predictions(
        dataset=dataset,
        predictor=predictor,
        num_samples=num_samples
    )
    # Convert iterators to lists to store all forecasts and time series in memory.
    forecasts = list(forecast_it)
    tss = list(ts_it)

    print("✅ Predictions generated successfully!")
    return forecasts, tss
```

```python
# --- Analysis of Results: Performance Metrics ---
def analyze_forecasts(forecasts, test_dataset_original_df, target_series_names, prediction_length):
    """
    Analyzes forecast performance by calculating MAE, RMSE, and MAPE for each target series.
    Compares the forecasted values to the actual values within the prediction horizon.

    Args:
        forecasts: List of Forecast objects generated by LagLlama.
        test_dataset_original_df: The original pandas DataFrame used to create the test_dataset.
                                  This is used to get accurate actual values for the prediction horizon.
        target_series_names: List of strings, the names of the series that were targeted for prediction.
        prediction_length: Integer, the number of time steps predicted.

    Returns:
        dict: A dictionary where keys are series names and values are dictionaries containing
              'actual', 'forecast_mean', 'forecast_lower', 'forecast_upper', and calculated metrics (mae, rmse, mape).
    """
    results = {}

    # Iterate through each forecast generated by the model.
    for forecast in forecasts:
        series_name = forecast.item_id # Get the name of the series from the forecast object.

        # Check if this series is one of the designated primary targets.
        if series_name in target_series_names:
            # Determine the start and end of the actual prediction horizon in the test data.
            # CRITICAL FIX: Convert forecast.start_date (Period) to Timestamp for slicing DatetimeIndex.
            # And also ensure the forecast start/end dates are timezone-naive to match the DataFrame index.
            forecast_start_date_ts = forecast.start_date.to_timestamp().tz_localize(None)
            forecast_end_date_ts = forecast_start_date_ts + pd.Timedelta(prediction_length - 1, unit=FREQ)

            # Extract actual values from the original test_dataset_original_df for the prediction horizon.
            # Using .loc for date-based slicing with Timestamp objects.
            actual_values_in_horizon = test_dataset_original_df.loc[forecast_start_date_ts : forecast_end_date_ts, series_name]

            # Ensure the lengths of actuals and forecasts match for accurate metric calculation.
            if len(actual_values_in_horizon) == len(forecast.mean):
                # Calculate Mean Absolute Error (MAE).
                mae = np.mean(np.abs(actual_values_in_horizon.values - forecast.mean))
                # Calculate Root Mean Squared Error (RMSE).
                rmse = np.sqrt(np.mean((actual_values_in_horizon.values - forecast.mean) ** 2))
                # Calculate Mean Absolute Percentage Error (MAPE).
                # Add a small epsilon to the denominator to avoid division by zero for zero actual values.
                mape = np.mean(np.abs((actual_values_in_horizon.values - forecast.mean) / (np.abs(actual_values_in_horizon.values) + 1e-8))) * 100

                # Store all relevant data and calculated metrics.
                results[series_name] = {
                    'actual': actual_values_in_horizon.values,
                    'forecast_mean': forecast.mean,
                    'forecast_lower': forecast.quantile(0.1),
                    'forecast_upper': forecast.quantile(0.9),
                    'mae': mae,
                    'rmse': rmse,
                    'mape': mape
                }
            else:
                print(f"Warning: Length mismatch for {series_name} forecast and actual values.")
                print(f"Actuals length: {len(actual_values_in_horizon)}, Forecast length: {len(forecast.mean)}. Skipping metrics.")
        # else:
        #     print(f"Skipping analysis for series '{series_name}' as it is not a primary target.")

    return results
```
